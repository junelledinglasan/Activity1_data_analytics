<h3>Question 8</h3>
<h3>
Do the three penguin species (Adelie, Chinstrap, Gentoo) significantly differ in terms of body mass?</h3>

```python
from scipy.stats import f_oneway
```

```python
stats, p_value = f_oneway(
    penguins[penguins['species'] == 'Adelie']['body_mass_g'],
    penguins[penguins['species'] == 'Chinstrap']['body_mass_g'],
    penguins[penguins['species'] == 'Gentoo']['body_mass_g']
)
```

```python
p_value
```

<h4>INSIGHT:</h4>
<h3>Body mass varies by specie, Gentoo are the heaviest, Adelie the lightest.</h3>

</h3>
